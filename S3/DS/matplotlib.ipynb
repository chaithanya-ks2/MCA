```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [4, 5, 6]

plt.plot(x, y, marker='o', color='green', ms=8, mec='red', mfc='white')
#marker size, marker edge color, marker face color
plt.show()
```

```python
x = [1.3, 2, 3, 4]
y = [4, 4.3, 6, 6.5]

plt.plot(x, y, 'o--r')
#marker line color
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 3.5, 7, 8]
z = [1.5, 2.5, 1.25]

plt.plot(x, marker='<')
plt.plot(y,marker='s')
plt.plot(z, marker='v')

plt.xlabel("x values")
plt.ylabel("y values")

plt.title("My first plot")
plt.grid(axis='x', linestyle='--', color='purple', linewidth=.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 6, 7, 8]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x)
plt.xlabel("x values")
plt.ylabel("y Values")
plt.grid(linestyle='--', linewidth=.5)


plt.subplot(1, 2, 2)
plt.plot(y)
plt.xlabel("x values")
plt.ylabel("y Values")

plt.subplot(2, 2, 1)
plt.plot(x)
plt.xlabel("x values")
plt.ylabel("y Values")
plt.grid(linestyle='--', linewidth=.5)

plt.subplot(2, 2, 2)
plt.plot(x)
plt.xlabel("x values")
plt.ylabel("y Values")
plt.grid(linestyle='--', linewidth=.5)

plt.suptitle("Subplotting sample")

plt.show()
```
