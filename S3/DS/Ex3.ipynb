```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
x = [1, 2, 6, 18]
y = [3, 10, 12, 20]

plt.plot(x, y, 'o:r', mfc='green', mec='green')
plt.show()
```

```python
temp = [12, 14, 16, 18, 20, 22, 24]
sales = [100, 200, 250, 400, 300, 450, 500]

plt.scatter(temp, sales)
plt.xlabel("Temperature (degree Celcius)")
plt.ylabel("Sales")
plt.title("Temperature vs Sales")

plt.show()
```

```python
list1 = []
list2 = []

with open('test.txt', 'r') as file:
    for line in file:
        parts = line.strip().split(',')
        list1.append(int(parts[0]))
        list2.append(int(parts[1]))

plt.plot(list1, list2, 'o--r')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Ploting line from text file")

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 3.5, 7, 8]
z = [1.5, 2.5, 1.25]

plt.plot(x, marker='<')
plt.plot(y,marker='s')
plt.plot(z, marker='v')

plt.xlabel("x values")
plt.ylabel("y values")

plt.title("Multiple plots in same plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]  

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y, marker='o')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plot 1")
plt.grid(linestyle='--', linewidth=0.5)

plt.subplot(1, 2, 2)
plt.plot(y, x, marker='s', color='orange')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plot 2")
plt.grid(linestyle='--', linewidth=0.5)

plt.tight_layout(rect=[0, 0, 1, 0.95])  
plt.suptitle("Subplotting Sample", fontsize=16)

plt.show()
```

```python
languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 77, 6.7]

plt.bar(languages, popularity)
plt.show()
```

```python
plt.barh(languages, popularity, color="red")
plt.show()
```

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan']

plt.bar(languages, popularity, color=colors)
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Programming Language Popularity')
plt.show()
```

```python
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

labels = ['A', 'B', 'C', 'D', 'E']
x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, men, width, color='green', label='Men')
plt.bar(x + width/2, women, width, color='red', label='Women')

plt.xlabel('Category')
plt.ylabel('Scores')
plt.title('Scores by Gender and Category')
plt.xticks(x, labels)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Programming Language Popularity')
plt.show()
```
